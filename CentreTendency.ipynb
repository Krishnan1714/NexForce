```python
import pandas as pd
path = "/content/drive/MyDrive/Dataset/SOCR-HeightWeight.csv"
fl = pd.read_csv(path)
Height_list =list(fl["Height(Inches)"])
n_h = len(Height_list)
Mean_Height = sum(Height_list)/n_h
Weight_list =list(fl["Weight(Pounds)"])
n_w = len(Weight_list)
Mean_Weight = sum(Weight_list)/n_w
print("Mean Height is {:.4f} Inches\nMean Weight is {:.4f} Pounds".format(Mean_Height,Mean_Weight))
```

```python
Height_list.sort()
Weight_list.sort()
if n_h%2==0:
    Median = (((n_h//2)-1)+(n_h//2))//2
else:
    Median = ((n_h+1)//2)-1
Median_Height = Height_list[Median]
Median_Weight = Weight_list[Median]
print("Median Height is {:.4f} Inches\nMedian Weight is {:.4f} Pounds".format(Median_Height,Median_Weight))
```

```python
import statistics as st
print("Mode Height is {:.4f} Inches".format(st.mode(Height_list)))
print("Mode Weight is {:.4f} Pounds".format(st.mode(Weight_list)))
```
